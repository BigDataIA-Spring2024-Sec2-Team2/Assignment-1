```python
from diagrams import Diagram, Cluster, Edge
from diagrams.azure.web import AppServiceDomains
from diagrams.aws.compute import Lambda
from diagrams.aws.database import Database
from diagrams.custom import Custom
from urllib.request import urlretrieve
from diagrams.aws.analytics import DataLakeResource
from diagrams.programming.language import Python

graph_attr = {
    "fontsize": "25",
    "bgcolor": "transparent",
    "width": "1000",
    "height": "1000"
}

cluster_attr = {
    "bgcolor": "white",
     "fontsize":"15"
}

with Diagram("\n\nData Architecture Diagram \nAirlines Service Review System", show=False, graph_attr=graph_attr) as diag:

      snowpipe_url = "https://community.snowflake.com/profilephoto/7293r000000gz0f/M"
      snowpipe_icon = "openstack.png"
      urlretrieve(snowpipe_url, snowpipe_icon)
      snowpipe = Custom("Snowpipe \n\nData ingestion", snowpipe_icon, edge_label=" ", fontsize="15", x_pos="1", y_pos="1", fontweight="bold")




      with Cluster("Data \nLoading ", graph_attr=cluster_attr):
        snowflake_db_url = "https://pbs.twimg.com/media/EDT-rsuWwAAmhav.png"
        nowflake_db_icon = "elastic.png"
        urlretrieve(snowflake_db_url, nowflake_db_icon)
        snowflake_db = Custom("Snowflake \nData Warehouse \n\nFinal data \ndestination", nowflake_db_icon, edge_label=" ", fontsize="15", x_pos="1", y_pos="1")

      website_icon = AppServiceDomains("airlinequality.com", fontsize="15", edge_label=" ", x_pos="1", y_pos="1")
      with Cluster("Data Extraction and Transformation", direction="TB", graph_attr=cluster_attr):
            with Cluster("Airflow DAGs", direction="TB", graph_attr=cluster_attr):
              python_extract = Python("\nWeb Scrapping \n(beautifulsoup)", fontsize="15")
              python_transform = Python("\nTransform \nHTML to Data", fontsize="15")
            data_lake = DataLakeResource("\nDatalake \n\nStoring raw \ndata in S3", fontsize="15", edge_label=" ", x_pos="1", y_pos="1")
            staging = Database("\nInternal Staging \n\nTemporary \nstorage", fontsize="15", edge_label=" ", x_pos="1", y_pos="1")



      website_icon >> python_extract >> Edge(label="Raw HTML") >> data_lake >> Edge(label="Raw HTML") >> python_transform >> Edge(label="Transformed Data") >> staging >> snowpipe >> snowflake_db
diag
```
